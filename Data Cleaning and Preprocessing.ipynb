# Deciding City 

```python
import os
import numpy as np
import pandas as pd

data = pd.read_csv("US_Accidents_March23.csv")
data.head()
```

```python
print(data.shape[0])
print(data.shape[1])
```

```python
city_counts = data["City"].value_counts()

very_large_cities_series = city_counts[city_counts > 20000]
very_large_data_cities2 = very_large_cities_series.index.tolist()

print("20.000'den fazla veriye sahip şehirler ve sayıları:")
print(very_large_cities_series)

for i, (city, count) in enumerate(very_large_cities_series.items()):
    print(f"Index: {i}")
    print(f"Şehir: {city}, Kayıt Sayısı: {count}")
```

```python
for city in very_large_data_cities2:
  city_df = data[data["City"] == city]
  print("Şehir: ",city)
  print("Toplam Data: ",city_df.shape[0])
  # 1. Kaza Şiddeti Dağılımı (Yüzdesel)
  print("\n1. Kaza Şiddeti Dağılımı:")
  severity_dist = city_df['Severity'].value_counts(normalize=True) * 100
  print(severity_dist.round(2).astype(str) + '%')

  # 2. En Sık Görülen Hava Durumları
  print("\n2. En Sık Görülen 10 Hava Durumu:")
  print(city_df['Weather_Condition'].value_counts())

  # 3. Yol Özelliklerinin Varlığı (Yüzdesel)
  print("\n3. Yol Özelliklerinin Kaza Anındaki Varlığı:")
  road_features = ['Junction', 'Stop', 'Traffic_Signal']
  for feature in road_features:
      feature_percentage = city_df[feature].mean() * 100
      print(f"   - {feature}: {feature_percentage:.2f}%")
```

# Atlanta

```python
atlanta_df = data[data["City"]=="Atlanta"].copy()
atlanta_df.head()
```

```python
print(atlanta_df.shape[0])
```

```python
# --- Eksik Veri Analizi ---
print("\n--- Atlanta Verisi İçin Eksik Veri Analizi ---")
missing_percent = (atlanta_df.isnull().sum() / len(atlanta_df)) * 100
missing_df = missing_percent[missing_percent > 0].sort_values(ascending=False)

if not missing_df.empty:
    print("Sütunlardaki eksik veri yüzdeleri:")
    print(missing_df.round(2).astype(str) + '%')
else:
    print("Harika! Seçilen sütunlarda hiç eksik veri bulunmuyor.")

# --- Veri Tiplerini Düzeltme ---
print("\n--- Veri Tiplerini Düzeltme ---")
atlanta_df['Start_Time'] = pd.to_datetime(atlanta_df['Start_Time'], errors='coerce')
atlanta_df['End_Time'] = pd.to_datetime(atlanta_df['End_Time'], errors='coerce')
print("'Start_Time' ve 'End_Time' sütunları datetime formatına çevrildi.")
```

```python
atlanta_df['Start_Time'] = pd.to_datetime(atlanta_df['Start_Time'], errors='coerce')
atlanta_df['End_Time'] = pd.to_datetime(atlanta_df['End_Time'], errors='coerce')

# --- ÖZELLİK MÜHENDİSLİĞİ (Adım 1: Zaman Grupları) ---
print("\n--- Adım 1: Doldurma işlemi için zaman grupları oluşturuluyor ---")
atlanta_df['Hour'] = atlanta_df['Start_Time'].dt.hour
atlanta_df['Month'] = atlanta_df['Start_Time'].dt.month
print("-> Hour ve Month sütunları eklendi.")
```

```python
atlanta_df['Start_Time'] = pd.to_datetime(atlanta_df['Start_Time'], errors='coerce')
atlanta_df['End_Time'] = pd.to_datetime(atlanta_df['End_Time'], errors='coerce')
atlanta_df.dropna(subset=['Start_Time', 'End_Time'], inplace=True)
print(f"Zaman verisi olmayan satırlar çıkarıldı. Kalan satır: {len(atlanta_df)}")
```

```python
import numpy as np

# --- AKILLI EKSİK VERİ YÖNETİMİ (Adım 2) ---
print("\n--- Adım 2: Grup Bazlı Akıllı Eksik Veri Yönetimi ---")

# Yüksek oranlı eksik veri içeren sütunları kaldır
cols_to_drop = ['End_Lat', 'End_Lng', 'Wind_Chill(F)', 'Precipitation(in)']
# !!! HATA DÜZELTİLDİ: df yerine atlanta_df kullanılıyor !!!
existing_cols_to_drop = [col for col in cols_to_drop if col in atlanta_df.columns]
atlanta_df.drop(columns=existing_cols_to_drop, inplace=True, errors='ignore')
print(f"Kaldırılan sütunlar: {existing_cols_to_drop}")

# Grup bazlı doldurma (Imputation)
numeric_cols_to_fill = ['Wind_Speed(mph)', 'Humidity(%)', 'Temperature(F)', 'Visibility(mi)', 'Pressure(in)']
print("Sayısal sütunlar grup ortalamaları ile dolduruluyor...")
for col in numeric_cols_to_fill:
    if col in atlanta_df.columns:
        atlanta_df[col] = atlanta_df.groupby(['Month', 'Hour'])[col].transform(lambda x: x.fillna(x.mean()))
        atlanta_df[col].fillna(atlanta_df[col].mean(), inplace=True)

categorical_cols_to_fill = ['Wind_Direction', 'Weather_Condition', 'Sunrise_Sunset']
print("Kategorik sütunlar grup modu ile dolduruluyor...")
for col in categorical_cols_to_fill:
    if col in atlanta_df.columns:
        atlanta_df[col] = atlanta_df.groupby(['Month', 'Hour'])[col].transform(lambda x: x.fillna(x.mode()[0] if not x.mode().empty else "UNKNOWN"))
        atlanta_df[col].fillna(atlanta_df[col].mode()[0], inplace=True)

if 'Street' in atlanta_df.columns:
    atlanta_df['Street'].fillna('UNKNOWN', inplace=True)

print("Akıllı doldurma işlemi tamamlandı.")
print(f"İşlem sonrası kalan eksik veri sayısı: {atlanta_df.isnull().sum().sum()}")


# --- ÖZELLİK MÜHENDİSLİĞİ (Adım 3: Kalanlar) ---
print("\n--- Adım 3: Kalan Özelliklerin Mühendisliği ---")
atlanta_df['Day'] = atlanta_df['Start_Time'].dt.day_name()
atlanta_df['Rush_Hour'] = np.where(atlanta_df['Hour'].between(7, 9) | atlanta_df['Hour'].between(16, 18), 1, 0)
print("-> Day ve Rush_Hour sütunları oluşturuldu.")
```

```python
# 1. Her kategori için tam liste oluşturalım
clear_conditions = ['Clear', 'Fair']
cloudy_conditions = ['Cloudy', 'Mostly Cloudy', 'Partly Cloudy', 'Scattered Clouds', 'Overcast']
rainy_conditions = ['Rain', 'Heavy Rain', 'Light Rain', 'Light Drizzle', 'Drizzle', 'Light Rain Showers', 'Light Rain Shower', 'Rain Shower', 'Rain Showers']
thunderstorm_conditions = ['Heavy Thunderstorms and Rain', 'Thunderstorms and Rain', 'Light Thunderstorms and Rain', 'Thunderstorm', 'Light Rain with Thunder', 'T-Storm', 'Thunder', 'Heavy T-Storm', 'Thunder in the Vicinity']
snow_icy_conditions = ['Light Freezing Rain', 'Light Snow', 'Snow', 'Light Ice Pellets', 'Wintry Mix']
low_visibility_conditions = ['Haze', 'Fog', 'Mist', 'Patches of Fog', 'Smoke', 'Shallow Fog', 'Drizzle and Fog']
windy_conditions = ['Squalls', 'Fair / Windy', 'Cloudy / Windy', 'Mostly Cloudy / Windy', 'Partly Cloudy / Windy', 'Light Rain / Windy', 'Heavy T-Storm / Windy', 'Thunder / Windy', 'T-Storm / Windy', 'Wintry Mix / Windy', 'Rain / Windy', 'Heavy Rain / Windy', 'Light Drizzle / Windy', 'Light Snow / Windy']

# 2. Çeviri sözlüğü oluşturalım
weather_map = {}
for item in clear_conditions: weather_map[item] = 'Clear'
for item in cloudy_conditions: weather_map[item] = 'Cloudy'
for item in rainy_conditions: weather_map[item] = 'Rainy'
for item in thunderstorm_conditions: weather_map[item] = 'Thunderstorm'
for item in snow_icy_conditions: weather_map[item] = 'Snowy_Icy'
for item in low_visibility_conditions: weather_map[item] = 'Low_Visibility'
for item in windy_conditions: weather_map[item] = 'Windy'

# 3. .map() fonksiyonu ile doğrudan eşleme yapalım
atlanta_df['Weather_Group'] = atlanta_df['Weather_Condition'].map(weather_map)
print("-> Weather_Group sütunu doğrudan eşleme ile oluşturuldu.")


unmapped_count = atlanta_df['Weather_Group'].isnull().sum()
atlanta_df = atlanta_df.dropna(subset=['Weather_Group'])
if unmapped_count > 0:
    print(f"UYARI: {unmapped_count} adet hava durumu eşleştirilemedi. Lütfen listeyi kontrol edin.")

print("\nOluşturulan Hava Gruplarının Dağılımı:")
print(atlanta_df['Weather_Group'].value_counts())
```

```python
atlanta_df["Severity"] = atlanta_df["Severity"].apply(lambda x: 0 if x < 3 else 1)
```

```python
atlanta_df.dropna(inplace=True)
print(f"\nSon temizlik sonrası kalan satır sayısı: {len(atlanta_df)}")
print(f"Tüm işlemler sonrası kalan eksik veri sayısı: {atlanta_df.isnull().sum().sum()}")

# 9. Kaydetme
atlanta_df.to_csv("atlanta_cleaned_data.csv", index=False)
```
